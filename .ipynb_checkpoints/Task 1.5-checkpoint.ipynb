### 1. Import Libraries

```python
# Import Libraries
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#import matplotlab 
import nltk
from nltk.corpus import stopwords
import re
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
#nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
from collections import Counter
sns.set()
```

### 2. Load 20th_century data

```python
# Import txt file
with open('20th_century_scrape.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

### 3. Tokenize words from text and create bar chart

```python
# Word tokenization
tokenized_word = word_tokenize(data)
print(tokenized_word[:100]) 
```

```python
# Create barchart of most common words
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
#print(dist_words) 
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot
chart_data = list(dist_words.most_common(10))
x, y = zip(*chart_data)  # splits the data pairs into two list x and y
plt.bar(x, y)
plt.show()
```

### 4. Remove stop words, and rerun plot and comment on what changed

```python
# Removing stopwords
stop_words = set(stopwords.words('english'))
# Add "The" ???? XXXXXX
stop_words.add("The")
filtered_words = [] # creates an empty list
for word in tokenized_word:
      if word not in stop_words:
           filtered_words.append(word)
```

```python
# Substitute all punctuation marks with a space
sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
tokenized2 = word_tokenize(sans_punc)
```

```python
# Noticed that s and th are showing up. Removing them.
words_to_remove = ["th", "s","www"]
for word in words_to_remove:
    while word in tokenized2:
        tokenized2.remove(word)
```

```python
dist_words = FreqDist(tokenized2)
```

```python
# Frequency Distribution Plot
chart_data = list(dist_words.most_common(10))
x, y = zip(*chart_data)  
plt.bar(x, y)
plt.show()
```

#### Q5 Comment: The new plot removed function words as well as punctuation and plotted the 10 most common words in the article.

### 5. Create a tags list using TextBlob

```python
# Creating tags list
text = TextBlob(" ".join(tokenized2))
```

```python
tags_list = text.tags
```

```python
tags_list[:50]
```

### 6. Create a list of the top 10 POS tags in article

```python
# Create top 10 POS tags
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top10=df_t.nlargest(10, 'Words')
top10
```

### 7. Plot list in a bar chart using the seaborn library

```python
# Plotting list in bar chart
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top10).set_title("20th_century - top 10 word types used")
```

### 8. Create three bar plots with the top 15 POS labels

```python
# Creating bar plot for "Nouns"
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences') 
top10
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x = "Occurrences", y = "Word", saturation = 0.9, data = top10).set_title("20th_century - Nouns")
```

```python
# Creating bar plot for "Verbs"
df = df_text[(df_text['Word type'] == "VBG") | (df_text['Word type'] == "VBD") | (df_text['Word type'] == "VBP")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences') 
top10
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x = "Occurrences", y = "Word", saturation = 0.9, data = top10).set_title("20th_century - Verbs")
```

```python
# Creating bar plot for "Adjectives"
df = df_text[(df_text['Word type'] == "JJ")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences') 
top10
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x = "Occurrences", y = "Word", saturation = 0.9, data = top10).set_title("20th_century - Adjectives")
```

### 9. Create a dataframe with the countries and the number of times they're mentioned in the text

```python
# Reading countries list from Task 1.3
with open('Country_Scrape.txt', 'r', errors='ignore') as file: 
    data = file.read().strip()

countries = data.split("\n")
```

```python
# Creating list of countries and frequency/count
countries_d = {}
for c in countries : 
    if c in dist_words:
        countries_d[c]=dist_words[c]
countries_d
```

```python
# Creating dataframe
countries_col = list(countries_d.keys())
freq_col = list(countries_d.values())
df_countries = pd.DataFrame()
df_countries['Country'] = countries_col
df_countries['Frequency'] = freq_col
df_countries
```

### 10. Create a plot showing the frequency of the number of times these countries are mentioned in the text

```python
# Creating plot showing frequency of countries in 20th century wiki page
plt.figure(figsize = (20, 10))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x = "Frequency", y = "Country", saturation = 0.9, data = df_countries).set_title("20th_century - Frequency of Countries")
```
